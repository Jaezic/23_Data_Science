```python
from pycaret.classification import *
import pandas as pd
data = pd.read_csv('./dataset/preprocessed.csv')

from sklearn.model_selection import KFold, train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.drop('scale_damage', axis=1), data['scale_damage'], test_size=0.2, random_state=123)
data = pd.concat([x_train, y_train], axis=1)
s = setup(data, target = 'scale_damage', session_id = 123)
```

```python
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
exp.setup(data, target = 'scale_damage', session_id = 123, normalize=True)
```

```python
models()
```

```python
best_5 =compare_models(fold=10, sort='Accuracy', n_select=5)
```

```python
df = pull()
```

```python
df
```

```python
top5 = [rank for rank in best_5]

tuned_top5 = [tune_model(i) for i in top5]
```

```python
blended = blend_models(estimator_list = tuned_top5,
                       fold = 10,
                       method = 'soft',
                       optimize='Accuracy',
                       )
```

```python
print(blended)
```

```python
final_model = finalize_model(blended)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, final_model.predict(x_test)))

from sklearn.metrics import classification_report
print(classification_report(y_test, final_model.predict(x_test)))

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, final_model.predict(x_test)))
```
